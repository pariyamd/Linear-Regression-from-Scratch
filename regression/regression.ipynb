```python
import pandas as pd;
import matplotlib.pyplot as plt;
import numpy as np;
import seaborn as sn;
```

```python
data=pd.read_csv("housing-prices/train.csv")
```

```python
x=data['LotArea']
```

```python
y=data['SalePrice']
```

```python
plt.figure(figsize=(20,10))
x=x/np.array(x).std()
y=y/np.array(y).std()
plt.scatter(x,y)
plt.show()
```

```python
theta=[0,4]
m=len(x)

def cost(x,y,theta,m):
    sigma=0
    for i in range(m):
        sigma+=(theta[0]+theta[1]*x[i]-y[i])**2
    return sigma/(2*m)
```

```python

    
def gradient_descent(x,y,theta,alpha,iteration,m):
    plt.scatter(x,y)
    xl=np.linspace(1,max(x))
    for t in range(iteration):
        t0=t1=0
        for i in range(m):
            t0+=(theta[0]+theta[1]*x[i]-y[i])
            t1+=(theta[0]+theta[1]*x[i]-y[i])*x[i]
        theta[0]=theta[0]-alpha*(t0)/m
        theta[1]=theta[1]-alpha*(t1)/m
        yl=theta[0]+theta[1]*xl
        plt.plot(xl,yl)
    return theta
        
```

```python
theta=gradient_descent(x,y,[0,10],0.500,200,m)
plt.show()
```

```python
plt.scatter(x,y)
xl=np.linspace(1,max(x))
yl=theta[0]+theta[1]*xl
plt.plot(xl,yl)
```
